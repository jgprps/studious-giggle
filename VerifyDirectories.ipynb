```
#VerifyDirectories version 2
import os, re

#open list of barcodes from spreadsheet copy/paste.

dosufile = open("barcodes.txt", 'r')
dosulist = [line.strip() for line in dosufile]
dosufile.close()

cleanbarcodes = []
downbarcodes  = []

for computer in dosulist:
  command  = 'ping -n 1 ' + computer 
  response = os.system(command)
  print(response)
  if response == 0:
    cleanbarcodes = cleanbarcodes + [computer]
  else:
    downbarcodes  = downbarcodes + [computer]





acuteCodes    = []
downtimeCodes = []
launchCodes   = []
secCodes      = []
ambCodes      = []
dtbCodes      = []
IcaRegex = re.compile('S.+\.lnk$') #establish a regular expression to identify ica files beginning with S

for computer in cleanbarcodes:
  
  curpath = '\\\%s\\C$\\Users\\Public\\Desktop' % computer
  

  #directory_list = 'dir ' + '"' + curpath + '"'
  #os.system(directory_list) # previous version - now using native python libraries instead of system call.

  #begin testing each computer's path.  If path exists, place each file of the directory into a directoryList

  if os.path.exists(curpath):
      print()
      print(computer)
      directoryList  = os.listdir(curpath)
      
      # Loop through every file in the directoryList and search for relevant strings. 
      
      for file in directoryList:
         print(file)
         if re.search('SCM Downtime', file):
            downtimeCodes = downtimeCodes + [computer]
            

         elif IcaRegex.search(file):
            if re.search('SEC', file):
              secCodes    = secCodes + [computer]
            elif re.search('Ambulatory', file):
              ambCodes    = ambCodes + [computer]
            elif re.search('Sunrise', file):
              acuteCodes  = acuteCodes + [computer]
              
            
            
         elif re.search('Launch SCM', file):
            launchCodes   = launchCodes + [computer]
            

         elif re.search('ED-DTB', file):
            dtbCodes      = dtbCodes + [computer]
            

  #if directory does not exist, flag as a 'bad path'.

  else:
      print() 
      print('bad path!'+ curpath)

#print a report of computers
print()
print("The barcodes that could not be accessed are:\n")
print(str(downbarcodes).strip('[]'))
print("The ED DTB computers are:\n")
print(str(dtbCodes).strip('[]'))
print("The Downtime computers are:\n")
print(str(downtimeCodes).strip('[]'))
print("The Launch SCM computers are:\n")
print(str(launchCodes).strip('[]'))
print("The SEC computers are:\n")
print(str(secCodes).strip('[]'))
print("The Ambulatory computers are:\n")
print(str(ambCodes).strip('[]'))
print("The Acute computers are:\n")
print(str(acuteCodes).strip('[]'))


```
